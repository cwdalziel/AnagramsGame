## Importing libraries

```python
from pytube import YouTube
from pydub import AudioSegment
from IPython.display import Audio
```

# Getting Audio File

```python
while True:
  url = input("Enter a YouTube URL: ")
  try:
    yt = YouTube(url)
    break
  except:
    print("Not a valid URL, try again")
file = yt.streams.get_audio_only().download()
```

# Pitch Shifting

## Getting user input

```python
while True:
  try:
    semitones = int(input("How many semitones should the audio be shifted?: "))
    break
  except:
    print("Invalid number of semitones, try again")
```

## Applying pitch shift

```python
audio = AudioSegment.from_file(file)
audio = audio._spawn(audio.raw_data, overrides={ "frame_rate": int(audio.frame_rate * (2 ** (1./12)) ** semitones)})
final = audio.export("/content/\"" + yt.title + "\"_" + str(semitones) + "_semitone_shift.mp3", format = "mp3")
```

# Finished File

```python
Audio(final.name, autoplay=True)
```
